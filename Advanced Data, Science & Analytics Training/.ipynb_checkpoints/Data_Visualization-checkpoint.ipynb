#  Matplotlib & Seaborn

Matplotlib and Seaborn are two of the most commonly used visualization libraries in Python. Understanding their differences and use cases helps in selecting the right tool for different types of data visualizations.

## Matplotlib

- Low-level plotting library for creating custom visualizations
- Requires more code to create visualizations
- Highly customizable but requires more effort
- Simple or highly customized plots, line charts, bar charts, and scatter plots

### When to use Matplotlib

- When you need full control over your plot (e.g., custom labels, legends, annotations).
- When working with non-statistical data that needs precise formatting.

## Seaborn

- High-level API built on Matplotlib, designed for statistical data visualization
- Provides built-in themes and functions for complex plots with minimal code
- Built-in themes and color palettes make it visually appealing by default
- Statistical plots such as histograms, boxplots, violin plots, and correlation heatmaps

### When to use Seaborn

- When working with statistical data and need quick insights.
- When visualizing relationships, distributions, and categorical data.

### Basic Plotting with Matplotlib

```python
import matplotlib.pyplot as plt # pyplot provides functions to create and customize plots easily
# Sample data

x = [10, 20, 30, 40, 50]
y = [100, 200, 250, 300, 50]

plt.plot(x,y,marker = 'o', linestyle = '-', color = 'b', label = "Sample Line")

#labels and title
# Labels and title
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Simple Line Plot")
plt.legend()

# Show the plot
plt.show()
```

### Key Matplotlib Functions

- plt.plot() - creates a lone graph
- plt.scatter() - creates a scatter plot
- plt.bar() -     created a bar chart
- plt.hist() - ctreates a histogram
- plt.xlabel() / plt.ylabel() - Sets labels for axes
- plt.title () - sets the title of the plot
- plt.legend() -  adds legends
- plt.show() - displayes the ply

1. Plot a line graph showing the relationship between years and population.

```python
# import matplotlib.pyplot as plt

years = [2000, 2005, 2010, 2015, 2020]
population = [2.5, 2.8, 3.0, 3.3, 3.7]

plt.plot(years, population, marker='o', linestyle='-', color='b', label="Population")

plt.xlabel("Year")
plt.ylabel("Population (millions)")
plt.title("Population Growth Over Years")
plt.legend()

plt.show()
```

2. A bar chart showing sales trends for five months.

```python
months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [500, 700, 800, 650, 900]

plt.bar(months, sales, color='green')

plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Report")

plt.show()
```

3. Create an interactive scatter plot using mplcursors.

```python
pip install mplcursors
```

```python
import matplotlib.pyplot as plt
import numpy as np
import mplcursors  # Allows interactive tooltips

# Generate random data
np.random.seed(42)
x = np.random.rand(50) * 100
y = np.random.rand(50) * 100

plt.scatter(x, y, color='purple', alpha=0.6, edgecolors='black')

plt.xlabel("X-axis Value")
plt.ylabel("Y-axis Value")
plt.title("Interactive Scatter Plot")

# Enable interactive tooltips
mplcursors.cursor(hover=True)

plt.show()
```

1. Create a scatter plot showing the relationship between total bill and tips.

```python
#import seaborn as sns
#import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.scatterplot(x="total_bill", y="tip", data=tips)

plt.title("Tip Amount vs Total Bill")

plt.show()
```

```python
sns.histplot(tips["total_bill"], bins=20, kde=True)

plt.title("Distribution of Total Bill")

plt.show()
```

3. Create a boxplot to compare tips across different days.

```python
sns.boxplot(x="day", y="tip", data=tips)

plt.title("Tips by Day")

plt.show()
```

5. Create a bar chart with customized colors, annotations, and a style theme.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
fruits = ["Apples", "Bananas", "Cherries", "Dates", "Elderberries"]
sales = [50, 75, 40, 85, 90]

# Set Seaborn style
sns.set_style("darkgrid")

# Create bar plot
ax = sns.barplot(x=fruits, y=sales, palette="coolwarm")

# Add annotations
for i, value in enumerate(sales):
    plt.text(i, value + 2, str(value), ha='center', fontsize=12, fontweight='bold')

plt.xlabel("Fruit Type")
plt.ylabel("Sales Volume")
plt.title("Fruit Sales Performance")

plt.show()
```

6. Create a correlation heatmap with advanced customizations 

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips") # Load dataset

tips_num = tips.select_dtypes(include = ['number'])
corr = tips_num.corr() 

corr = tips_num.corr() # Compute correlation matrix

# Create heatmap with customizations
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5, annot_kws={"size":12})

plt.title("Enhanced Correlation Heatmap")
plt.show()
```

7. Create a pairplot to explore relationships between multiple variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris") # Load dataset

sns.pairplot(iris, hue="species", palette="husl") # Create pairplot

plt.show()
```
